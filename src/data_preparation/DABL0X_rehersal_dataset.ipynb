```python
import pandas as pd
path = '../../data/training_data/DFINAL_VTrain.csv'
path_rehearsal_data = '../../data/training_data/warm-up_900samples.csv'
```

```python
df = pd.read_csv(path, index_col=0)
df_reh = pd.read_csv(path_rehearsal_data, index_col=0)
```

```python
df.columns
```

```python
for column in ['game', 'benchmark_version', 'game_id', 'model', 'experiment',
               'episode', 'Aborted', 'Lose', 'Success', 'text']:
    df_reh[column] = None
df_reh.drop('id', axis=1, inplace=True)
```

```python
df_reh.columns
```

```python
df_reh = df_reh.sample(frac=1, random_state=7331).reset_index(drop=True)
```

```python
df_reh.head()
```

```python
mixed_rows = []
index_reh = 0
for index in range(len(df)):
    # Add a row from df1
    mixed_rows.append(df.iloc[index])
    # Add a row from df2 (if available)
    if index%3 == 0 and index_reh < len(df_reh):
        mixed_rows.append(df_reh.iloc[index_reh])
        index_reh += 1

# Convert the mixed rows back to a DataFrame
mixed_df = pd.DataFrame(mixed_rows)

print(mixed_df)
```
