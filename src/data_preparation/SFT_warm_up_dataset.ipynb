```python
import random
import math
import json
import pandas as pd
import numpy as np
from itertools import product, permutations
```

```python
def add_letter_feedback(input: str, letter_feedback: list)-> str:
    output = ""
    for letter in input:
        n = random.randint(0, len(letter_feedback)-1)
        output += letter + letter_feedback[n]
    return output


def process_letter_feedback(string: str) -> str:
    splits = string.split()
    output = ""
    for i, element in enumerate(splits):
        if i < len(splits)-1:
            if splits[i+1] == "<true>":
                output += element
            elif splits[i+1] == "<false>":
                output += "<UNK>"
    return output

def partition(list, n):
    """Make n partitions from shuffeled list"""
    list_in = list
    random.shuffle(list_in)
    return [list_in[i::n] for i in range(n)]
```

```python
# all variatons
random.seed(0)
np.random.seed(0)

with open("../../data/nouns.json") as f:
    nounjson = json.load(f)
nouns = nounjson["data"]
words = random.sample(nouns, 200)
consonants = set("bcdfghjklmnpqrstvwxyz")
vowels = set("aeiou")
reverse_words = [string[::-1] for string in words]
upper_words = [word.upper() for word in words]
len_words = [str(len(word)) for word in words]
n_consonants = [sum(1 for c in word if c in consonants) for word in words]
n_vowels = [sum(1 for v in word if v in vowels) for word in words]
four_letters = [word for word in words if len(word) == 4]
five_letters = [word for word in words if len(word) == 5]
six_letters = [word for word in words if len(word) == 6]
seven_letters = [word for word in words if len(word) == 7]
words_no_consonants = [''.join([l for l in word if l not in consonants]) for word in words]
words_no_vowels = [''.join([l for l in word if l not in vowels]) for word in words]
words_first_letters = [word[0] for word in words]
words_last_letters = [word[-1] for word in words]
scrambled_words = [''.join(random.sample(word, len(word))) for word in words]
sorted_words = [''.join(sorted(word)) for word in words]
double_letter_words = [''.join(np.repeat(list(word),2)) for word in words]
triple_words = [' '.join(np.repeat(word, 3)) for word in words]

list_size = 5
list_words = words.copy()
word_lists = partition(list_words, math.floor(len(list_words)/list_size))
shuffeled_word_lists = [random.sample(wlist, len(wlist)) for wlist in word_lists]
first_choices = [w_l[0] for w_l in shuffeled_word_lists]
second_choices = [w_l[1] for w_l in shuffeled_word_lists]
rests = [w_l[2:] for w_l in shuffeled_word_lists]


positions = ["first", "second", "third"]
position_options = list(permutations(positions))


numbers_a = np.random.choice(range(20), size=len(words), replace=True)
#np.array(random.sample(range(20), len(words)))
numbers_a_str = [str(num) for num in numbers_a]
numbers_b = np.random.choice(range(30), size=len(words), replace=True)
#np.array(random.sample(range(30), len(words)))
numbers_b_str = [str(num) for num in numbers_b]
sum_ab = [str(sum) for sum in numbers_a + numbers_b]

large_numbers = [str(num + 100) for num in random.sample(range(10000), 200)]
reverse_numbers = [string[::-1] for string in large_numbers]

nonanagrams = ['rhythm', 'sphinx', 'lymph', 'vortex', 'zephyr', 'fjord', 'quartz', 'waltz', 'jinx', 'pyramid', 'galaxy', 'laptop', 'jovial', 'zebra', 'pumpkin', 'cactus', 'oxygen', 'banjo', 'nimbus', 'nymph', 'crypt', 'glyph', 'sylph', 'pygmy', 'flyby', 'calyx', 'quick', 'burqa', 'dwarf', 'fjeld', 'gowks', 'tizzy', 'phlox', 'query', 'ritzy', 'vivid', 'focus', 'funny', 'hobby', 'humid', 'ivory', 'juicy', 'kitty', 'livid', 'savvy', 'virus']
scrambled_nonanagrams = [''.join(random.sample(ele, len(ele))) for ele in nonanagrams]
nonanagrams_first_letters = [word[0] for word in nonanagrams]
nonanagrams_last_letters = [word[-1] for word in nonanagrams]


actions = ["water the plant", "break the tv", "vacuum the carpet", "buy the book", "search a game", "show the briefcase", "pull the lever", "light the candle", "slice the cucumber", "move your body", "build a fort", "invent a machine", "borrow some flour from the neighbour", "pet the pigeon", "greet the barista", "write the paragraph", "season the fish", "catch the mouse", "call your mother", "befriend the librarian", "watch the cloud", "burst the bubble", "paint the shed", "eat the watermelon", "mind the gap", "throw the ball", "order the cake", "sew the collar", "print the label", "cut the cord", "create a portal", "freeze the apricot", "pinch the seam", "press the button", "close the door", "save the elephant"]
action_groups = [[action] + random.sample(set(actions)-set([action]), 3) for action in actions]
action_topics = [action.split()[-1] for action in actions]

letter_feedback = [" <true> ", " <false> "]
letter_feedbacks = [add_letter_feedback(word, letter_feedback) for word in words]
processed_letter_feedbacks = [process_letter_feedback(input) for input in letter_feedbacks]

group_size = 2
part_words = words.copy()
word_groups = partition(part_words, math.floor(len(part_words)/group_size)) #might produce at least one partition that is larger than n, if the partition is not perfect
words_groups_perfect = [group for group in word_groups if len(group) == group_size]
#k:v
# a = {key: value for key, value in zip(words, words_first_letters)}
formatted1 = [list(zip(group, [word[0] for word in group])) for group in words_groups_perfect]
json1 = [{key: value for key, value in group} for group in formatted1]
json1 = [str(i) for i in json1]
#k:(v,v)
formatted2 = [list(zip(group, [(word[0], word[-1]) for word in group])) for group in words_groups_perfect]
json2 = [{key: value for key, value in group} for group in formatted2]
json2 = [str(i) for i in json2]
# #k: {k:v}
# c = {key: {value: number} for key, value , number in zip(words, words_first_letters, numbers_a)}
# # k:{k:[v]}
formatted3 = [[(word, word[0], random.randint(1, 100), random.randint(1, 50)) for word in group] for group in words_groups_perfect]
json3 = [{word: {letter : [int1, int2]} for word, letter, int1, int2 in group} for group in formatted3 ]
json3 = [str(i) for i in json3]

keywords = ["command", "ACTION", "Guess", "clue", "INSTRUCTION", "Aside", "reply", "CHOICE", "Pick", "solution", "ANSWER"] #"description", "explanation",
cap_keywords = [string.upper() for string in keywords]
title_keywords = [string.title() for string in keywords]
all_keywords = keywords + cap_keywords + title_keywords
```

```python
# generate data
chatlist = []
idlist = []
for minigame in minigames:
    #print(minigame["id"])
    #full_text = "[USER:] " + minigame["instruction"] + " [ASSISTANT:] " + minigame["answer"]
    full_text = "[{\"role\": \"user\", \"content\": \"" + minigame["instruction"] + "\"}, {\"role\": \"assistant\", \"content\": \"" + minigame["answer"] + "\"}]"

    # all possible combinations
    # TODO: check if lists are empty?
    if minigame["combination"] == "free":
        slotfillers = list(product(*minigame["replacements"].values()))
    elif minigame["combination"] == "fixed":
        slotfillers = list(zip(*minigame["replacements"].values()))
    elif minigame["combination"] == "mixed":
        replacements = minigame["replacements"]
        replacement_keys_non_fixed = [ x for x in list(replacements.keys()) if x not in minigame["sets"]]
        replacements_non_fixed = [replacements[key] for key in replacement_keys_non_fixed]
        
        pairings = list(zip(*([replacements[key] for key in minigame["sets"]])))
        slotfillers = []
        for pairing in pairings:
            #print(pairing)
            li = list(map(lambda el:[el], pairing))
            #print(li)
            slotfillers +=(list(product(*li, *replacements_non_fixed)))


    for slot in slotfillers:
        keys = minigame["replacements"].keys()
        replacement_dict = dict(zip(keys, list(slot)))
        #replacements = str.maketrans(replacement_dict)
        #res = knwo["instruction"].translate(replacements)
        string = full_text
        for k, v in replacement_dict.items():
            string = string.replace(k, str(v))
        #print(string)
        chatlist.append(string)
        idlist.append(minigame["id"])
        
    #print("-------------------------------------")
```

```python
#save data
df = pd.DataFrame({"id": idlist, "chat": chatlist})
df.to_csv("../../data/training_data/warm-up_all.csv")
df.head()
df.id.value_counts()
```
