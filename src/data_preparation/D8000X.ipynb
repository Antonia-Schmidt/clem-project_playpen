```python
path_image_game: str = '../../data/training_data/D30001.csv'
path_private_shared: str = '../../data/training_data/D20003.csv'
path_reference_game: str = '../../data/training_data/D20002.csv'
path_taboo: str = '../../data/training_data/D30003.csv'
path_wordle: str = '../../data/training_data/D30001.csv'
path_wordle_withclue: str = '../../data/training_data/D20001.csv'
wordle_withcritic: str = '../../data/training_data/D30001.csv'
```

```python
d30001 = pd.read_csv(path_image_game)
d20003 = pd.read_csv(path_private_shared)
d20002 = pd.read_csv(path_reference_game)
d30003 = pd.read_csv(path_taboo)
d20001 = pd.read_csv(path_wordle_withclue)
```

```python
df30001 = d30001[d30001.game.isin(['imagegame', 'wordle', 'wordle_withcritic'])]
df20003 = d20003[d20003.game.isin(['privateshared'])]
df20002 = d20002[d20002.game.isin(['referencegame'])]
df20001 = d20001[d20001.game.isin(['wordle_withclue'])]
df30003 = d30003[d30003.game.isin(['taboo'])]

df = pd.concat([df30001, df20003, df20002, df20001, df30003])
```
