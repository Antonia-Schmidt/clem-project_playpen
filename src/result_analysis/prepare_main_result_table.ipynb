```python
id_columns = [
      'imagegame % Played',
      'imagegame Quality Score',
      'privateshared % Played',
      'privateshared Quality Score',
      'referencegame % Played',
      'referencegame Quality Score',
      'taboo % Played',
      'taboo Quality Score',
      'wordle % Played',
      'wordle Quality Score',
      'wordle_withclue % Played',
      'wordle_withclue Quality Score',
      'wordle_withcritic % Played',
      'wordle_withcritic Quality Score'
]

ood_columns = [
    'adventuregame % Played',
    'adventuregame Quality Score',
    'codenames % Played',
    'codenames Quality Score',
    'guesswhat % Played',
    'guesswhat Quality Score',
    'matchit_ascii % Played',
    'matchit_ascii Quality Score',
    'textmapworld % Played',
    'textmapworld Quality Score',
    'textmapworld_graphreasoning % Played',
    'textmapworld_graphreasoning Quality Score',
    'textmapworld_specificroom % Played',
    'textmapworld_specificroom Quality Score',
]

baseline_lookup = {
  'L3-8B-0.7K': 'llama-8B-baseline',
  'L3-8B-1.1K': 'llama-8B-baseline',
  'L3-8B-1.7K': 'llama-8B-baseline',
  'L3-70B-0.6K': 'llama-70B-baseline',
  'L3-70B-1.1K': 'llama-70B-baseline',
  'L3-70B-1.7K': 'llama-70B-baseline',
  'MS-0.6K': 'mistral-small-baseline',
  'MS-1.1K': 'mistral-small-baseline',
  'MS-1.7K': 'mistral-small-baseline',
}

model_order = [
    #'llama-8B-baseline',
    'L3-8B-0.7K',
    'L3-8B-1.1K',
    'L3-8B-1.7K',
    #'mistral-small-baseline',
    #'MS-0.6K',
    #'MS-1.1K',
    #'MS-1.7K',
    ##'llama-70B-baseline',
    #'L3-70B-0.6K',
    #'L3-70B-1.1K',
    #'L3-70B-1.7K',
]
```

```python
def get_custom_score(row: dict, columns: list) -> float:
    played: int = 0
    count: int = 0
    for column in columns:
        p = row[column]
        if p >= 0:
            played += int(p)
            count += 1
        else:
            pass
    return played/count

def get_custom_clem_score(played: float, quality_score: float):
    return (played * quality_score) / 100
```

```python
import math


def get_color_coding(final_score: float, base_score: float) -> float:
    diff = final_score - base_score
    if diff < 0:
        return f'{final_score:.2f}' +  '{\\scriptsize \\textcolor{red}{\\rule{' + f'{math.fabs(diff): .2f}' + '}{5pt}}}'
    elif diff > 0:
        return f'{final_score:.2f}' +  '{\\scriptsize \\textcolor{green!60!black}{\\rule{' + f'{diff: .2f}' + '}{5pt}}}'

def get_baseline_row(model: str):
    d = df[df['model'] == model].iloc[0]
    v2_clemscore = d['clemscore']

    id_played = get_custom_score(d, [c for c in id_columns if '%' in c])
    id_quality = get_custom_score(d, [c for c in id_columns if '%' not in c])
    id_clem = get_custom_clem_score(id_played, id_quality)

    ood_played = get_custom_score(d, [c for c in ood_columns if '%' in c])
    ood_quality = get_custom_score(d, [c for c in ood_columns if '%' not in c])
    ood_clem = get_custom_clem_score(ood_played, ood_quality)

    return f"Baseline & {v2_clemscore:.2f} & {id_clem:.2f} & {id_played: .2f} & {id_quality:.2f} & {ood_clem:.2f} &  {ood_played: .2f} & {ood_quality:.2f} \\\\"

def get_colored_row(model: str, base_model: str):
    d = df[df['model'] == model].iloc[0]
    base = df[df['model'] == base_model].iloc[0]

    final_v2_clemscore = d['clemscore']
    final_id_played = get_custom_score(d, [c for c in id_columns if '%' in c])
    final_id_quality = get_custom_score(d, [c for c in id_columns if '%' not in c])
    final_id_clem = get_custom_clem_score(final_id_played, final_id_quality)
    final_ood_played = get_custom_score(d, [c for c in ood_columns if '%' in c])
    final_ood_quality = get_custom_score(d, [c for c in ood_columns if '%' not in c])
    final_ood_clem = get_custom_clem_score(final_ood_played, final_ood_quality)

    base_v2_clemscore = base['clemscore']
    base_id_played = get_custom_score(base, [c for c in id_columns if '%' in c])
    base_id_quality = get_custom_score(base, [c for c in id_columns if '%' not in c])
    base_id_clem = get_custom_clem_score(base_id_played, base_id_quality)
    base_ood_played = get_custom_score(base, [c for c in ood_columns if '%' in c])
    base_ood_quality = get_custom_score(base, [c for c in ood_columns if '%' not in c])
    base_ood_clem = get_custom_clem_score(base_ood_played, base_ood_quality)

    return f"{model} & {get_color_coding(final_v2_clemscore, base_v2_clemscore)} & {get_color_coding(final_id_clem, base_id_clem)} & {get_color_coding(final_id_played, base_id_played)} & {get_color_coding(final_id_quality, base_id_quality)} & {get_color_coding(final_ood_clem, base_ood_clem)} &  {get_color_coding(final_ood_played, base_ood_played)} & {get_color_coding(final_ood_quality, base_ood_quality)} \\\\"

for model in model_order:
    if model in baseline_lookup.keys():
        base_model = baseline_lookup[model]
        print(get_colored_row(model, base_model))
    else:
        base_model = model
        print(get_baseline_row(model))
```
