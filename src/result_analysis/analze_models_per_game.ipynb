```python
biggest_models = ["gpt-4-", "o1-preview-", "claude-3-5-", "gpt-4o-"]

def transform_model_name(name: str) -> str:
    if name.startswith("gpt-4-t"):
        return "GPT-4-Turbo"
    if name.startswith("gpt-4-"):
        return "GPT-4"
    elif name.startswith("o1-preview-"):
        return "O1-Preview"
    elif name.startswith("claude-3-5-"):
        return "Claude-3-5"
    elif name.startswith("gpt-4o-"):
        return "GPT-4o"
    elif name.startswith("claude-3-opus"):
        return "Claude-3-OPUS"
    else:
        return "other"


df["new_model"] = df.model.apply(transform_model_name)
data = (df.new_model.value_counts()/len(df))*100
```
